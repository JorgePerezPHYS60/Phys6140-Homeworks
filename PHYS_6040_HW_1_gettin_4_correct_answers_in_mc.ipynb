```python
from matplotlib import pyplot
import random

nquestions = 12
nchoices = 5

counts = [pow(10,i) for i in range(7)]
probs = []

for l in counts:
  count = 0
  for i in range(l):
    count_correct = 0
    for t in range(nquestions):
      if random.uniform(0,1) < 1.0/nchoices:
        count_correct += 1

    if count_correct <= 4:
      count += 1

  probs.append(count/l)


pyplot.subplot(2,1,1)
pyplot.plot(counts, probs, color="blue", lw=2)
pyplot.xscale('log')
pyplot.ylim( [0.8,1] )
pyplot.axline((0,0.927),(counts[len(counts)-1], 0.927))

pyplot.xlabel('Number of Trials')
pyplot.ylabel('Probability')
pyplot.show()

```
